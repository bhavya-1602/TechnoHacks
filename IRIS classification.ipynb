# Iris Flower Classification ML project

## Importing the liabraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

## Loading the dataset

```python
columns=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width','Class_labels']
#importing Dataset
df=pd.read_csv('iris.data',names=columns)
df
```

## Describing the dataset

```python
df.describe()
```

## Visualization of our dataset

```python
sns.pairplot(df, hue='Class_labels')
```

## Seperating Independent and Dependent Variables

```python
data=df.values

x=data[:,0:4]
y=data[:,4]
```

```python
print(x)
```

```python
print(y, y.shape)
```

## Splitting the Dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.8)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
print(x_test)
```

```python
print(y_test)
```

## MODEL SELECTION 

### Model 1 - Support Vector Machine (SVM)

```python
#Support Vector Machine
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(x_train,y_train)
```

```python
pred1=model_svc.predict(x_test)

#Calculate Accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pred1)*100)
```

### Model 2 - Logistic Regression

```python
#Logistic Regression Model
from sklearn.linear_model import LogisticRegression

model_LR = LogisticRegression()
model_LR.fit(x_train,y_train)
```

```python
pred2=model_LR.predict(x_test)

#Calculate accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pred2)*100)
for i in range (len(pred1)):
    print(y_test[i],pred1[i])
```

### Model 3 - Decision Tree Classifier

```python
#Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
model_DTC = DecisionTreeClassifier()
model_DTC.fit(x_train,y_train)
```

```python
pred3=model_DTC.predict(x_test)

#Calculate Accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pred3)*100)
for i in range(len(pred3)):
    print(y_test[i],pred3[i])
```

# Conclusion
### We apply the following 3 Classification models of machine learning on our given Iris Dataset:-

* Decision Tree Classifier
